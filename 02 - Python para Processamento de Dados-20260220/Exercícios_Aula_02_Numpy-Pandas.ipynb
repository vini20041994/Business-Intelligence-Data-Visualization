### Carregar EmendasParlamentares.csv e responder perguntas

```python
import pandas as pd
from pathlib import Path

# Definir caminho do arquivo
p = Path('EmendasParlamentares.csv')
if not p.exists():
    p = Path('Aula 02 - Python para Processamento de Dados-20260220') / 'EmendasParlamentares.csv'

# Carregar o arquivo em um dataframe (leitura básica)
df = pd.read_csv(p, sep=';', encoding='latin-1', engine='python')
df.columns = [c.strip() for c in df.columns]

print(f"Arquivo carregado: {p}")
print(f"\nDimensões do dataframe: {df.shape[0]} linhas × {df.shape[1]} colunas")
print(f"\nColunas: {df.columns.tolist()}")
print(f"\nPrimeiras 5 linhas:")
print(df.head())
```

### Preparação e conversão de dados
A seguir fazemos ajustes iniciais no `DataFrame`, formatando números e definindo colunas auxiliares.

```python
# ajustes e conversões numéricas (inspirado em exemplo_pandas_csv.py e Python for Data Analysis)

pd.options.display.float_format = '{:,.2f}'.format

def to_num(col):
    return pd.to_numeric(col.str.replace('.', '', regex=False).str.replace(',', '.', regex=False))

numeric_cols = ['Valor Empenhado', 'Valor Liquidado', 'Valor Pago', 'Valor Restos A Pagar Cancelados']
for c in numeric_cols:
    df[c + ' Num'] = to_num(df[c])

print("Conversões numéricas aplicadas às colunas relevantes.")
```

## Questões de análise
A seguir respondemos às perguntas solicitadas utilizando o DataFrame preparado.

### Q1: Valor total pago por ano

```python
# Q1: Qual o valor total pago em emendas por ano?
# coluna 'Valor Pago Num' já foi pré-calculada no carregamento (veja a célula inicial).
total_pago_por_ano = df.groupby('Ano da Emenda')['Valor Pago Num'].sum().sort_index()
print("\nValor total pago em emendas por ano:")
print(total_pago_por_ano)
print(f"\nTotal geral: R$ {total_pago_por_ano.sum():,.2f}")
```

### Q2: Média e desvio por região

```python
# Q2: Média e desvio padrão dos valores empenhados por região
# usamos o campo convertido anteriormente
q2_stats = df.groupby('Região')['Valor Empenhado Num'].agg(['mean', 'std', 'count', 'sum'])
print("Questão 2: Média e desvio padrão dos valores empenhados por região")
print(q2_stats)
```

### Q3: Top 10 autores por valor empenhado

```python
# Q3: Top 10 autores que mais destinaram recursos
q3_autores = df.groupby('Nome do Autor da Emenda')['Valor Empenhado Num'].sum().sort_values(ascending=False).head(10)
print("Questão 3: Top 10 autores que mais destinaram recursos (Valor Empenhado)")
print(q3_autores)
```

### Q4: Emendas para SC

```python
# Q4: Emendas para Santa Catarina
# filtro usando o mesmo valor que aparece no script de exemplo ('SANTA CATARINA')
q4_sc = df[df['UF'] == 'SANTA CATARINA'].shape[0]
q4_sc_valor = df[df['UF'] == 'SANTA CATARINA']['Valor Empenhado Num'].sum()
print(f"Questão 4: Emendas destinadas para Santa Catarina")
print(f"Total de emendas: {q4_sc}")
print(f"Valor total empenhado: R$ {q4_sc_valor:,.2f}")
```

### Q5: Municípios SC top por tipo

```python
# Q5: Municípios de SC que se destacam por tipo de emenda
# reutiliza o filtro por Estado que funcionou em Q4
# (o script de exemplo também cria `emendas_sc_2025` com UF == 'SANTA CATARINA')
df_sc = df[df['UF'] == 'SANTA CATARINA']
q5_municipios = df_sc.groupby(['Município', 'Tipo de Emenda'])['Valor Empenhado Num'].sum().sort_values(ascending=False).head(10)
print("Questão 5: Top 10 - Municípios de SC por tipo de emenda (Valor Empenhado)")
print(q5_municipios)
```

### Q6: Diferença Liquidado vs Pago

```python
# Q6: Emendas com diferença entre Valor Liquidado e Valor Pago
# os campos numéricos já estão preparados no carregamento
df['Diferença'] = abs(df['Valor Liquidado Num'] - df['Valor Pago Num'])
diferenca_significativa = df[df['Diferença'] > 0]
print(f"Questão 6: Emendas com diferença entre Valor Liquidado e Valor Pago")
print(f"Total de emendas com diferença > 0: {len(diferenca_significativa)}")
print(f"Percentual: {len(diferenca_significativa)/len(df)*100:.2f}%")
print(f"\nTop 10 maiores diferenças:")
print(diferenca_significativa.nlargest(10, 'Diferença')[['Ano da Emenda', 'Município', 'Valor Liquidado Num', 'Valor Pago Num', 'Diferença']])
```

### Q7: % Restos a Pagar Cancelados

```python
# Q7: Percentual de Restos a Pagar Cancelados por ano
# utilizando colunas numéricas previamente convertidas
q7_percentual = (df.groupby('Ano da Emenda')['Valor Restos A Pagar Cancelados Num'].sum() / df.groupby('Ano da Emenda')['Valor Empenhado Num'].sum()) * 100
print("Questão 7: Percentual de Restos a Pagar Cancelados em relação ao total empenhado por ano")
print(q7_percentual)
```

### Q8: Subfunção comum por região

```python
# Q8: Subfunção mais comum por região
q8_subfuncao = df.groupby('Região')['Nome Subfunção'].apply(lambda x: x.value_counts().index[0])
print("Questão 8: Subfunção mais comum para cada Região")
print(q8_subfuncao)
```

### Q9: Código IBGE ausente

```python
# Q9: Código Município IBGE ausente
q9_ausentes = df[df['Código Município IBGE'].isna() | (df['Código Município IBGE'] == '')]
q9_count = len(q9_ausentes)
q9_pct = (q9_count / len(df)) * 100
print(f"Questão 9: Código Município IBGE ausente")
print(f"Total de linhas com código ausente: {q9_count}")
print(f"Percentual: {q9_pct:.2f}%")
print(f"\nAmostra de linhas com código ausente:")
print(q9_ausentes[['Ano da Emenda', 'Município', 'Código Município IBGE', 'Localidade de aplicação do recurso']].head(10))
```

### Q10: Duplicatas de municípios

```python
# Q10: Duplicatas de Município com grafias diferentes
from difflib import SequenceMatcher
municipios_unicos = df['Município'].unique()
print(f"Questão 10: Análise de duplicatas de Município com grafias diferentes")
print(f"Total de nomes únicos de Município: {len(municipios_unicos)}")
print(f"Total de registros: {len(df)}")

# Exemplo: buscar municípios que começam com "São" ou variações
print(f"\nExemplos de Municípios com 'São' (primeiras 20):")
sao_municipios = [m for m in municipios_unicos if 'São' in m or 'Sao' in m or 'SAO' in m]
print(sao_municipios[:20])
```
